```python
import nltk
import pandas as pd
import re
from nltk import word_tokenize
import string 
from nltk.corpus import stopwords
import pickle


# read data
Location = r'/home/stratos/Desktop/ted_Project/twitter_data/test2017.tsv'
df = pd.read_csv(Location, error_bad_lines=False,
                 names=['no1','no2','cent','tweet'],sep='\t')

# remove punctuation
def strip_punctuation(s):
    return ''.join(c for c in s if c not in string.punctuation)

# remove links and make tokens
tokens=[]
for index, row in df.iterrows():
    row['tweet'] = re.sub(r"http\S+", "", row['tweet'])
    tok=nltk.word_tokenize(strip_punctuation(row['tweet'].lower()))
    tmp=[]
    for word in tok:
        if word in stopwords.words('english'):
            tmp.append(word)
    l3 = [x for x in tok if x not in tmp]
    if l3:
        tokens.append([l3,row['cent']])

print(tokens)

# save tokens
outfile=open('filtered_tweets_test.pickle','wb')
pickle.dump(tokens,outfile)
outfile.close()

# read tokens
infile = open('filtered_tweets_test.pickle','rb')
tokens = pickle.load(infile)
infile.close()

print(tokens)


```

```python
from collections import Counter

no_first_words=50

# # tockens=[[],[]...] filtered=[...]
filtered=[]
for tweet in tokens:
    filtered.extend(tweet)
    
# # find 30 most used words
count = Counter(filtered)
most_common=count.most_common(no_first_words)
print(most_common)

# plot histogram
df = pd.DataFrame.from_dict(dict(most_common), orient='index')
%matplotlib inline
df.plot(kind='barh',grid=True,figsize=(no_first_words/3,no_first_words/3),legend=False)
```
