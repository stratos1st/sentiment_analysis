```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.ensemble import RandomForestClassifier
from sklearn import neighbors
from sklearn.neighbors import NearestNeighbors
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from xgboost import XGBClassifier

import pandas as pd
import pickle
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
import numpy    
    
# read tokens
infile = open('lemmatized_tweets.pickle','rb')
tokens = pickle.load(infile)
infile.close()

# # concatinate tokens. ending up with a filtered tweet
tok=[]
for tweet in tokens:
    tmp=''
    for word in tweet:
        tmp=tmp+' '+word
    tok.append(tmp)
tokens=tok

# read data
Location = r'/home/stratos/Desktop/ted_Project/twitter_data/train2017.tsv'
df = pd.read_csv(Location, error_bad_lines=False,
                 names=['no1','no2','cent','tweet'],sep='\t')
df=df[:-1]
# ytrain=df.drop(columns=['no1','no2','tweet'])

test_vectori=[
    [CountVectorizer(max_features=3000)],
    [TfidfVectorizer(max_features=3000)]
]

test=[
    [neighbors.KNeighborsClassifier(n_jobs=4,algorithm='brute',weights='distance')],
    [LogisticRegression(n_jobs=4)],
    [RandomForestClassifier(n_jobs=4,n_estimators=250)],
    [svm.SVC(kernel='linear', C=1, probability=True)],
    [XGBClassifier()]
]


test1=[x+test_vectori[0] for x in test]
test2=[x+test_vectori[1] for x in test]

test=test1+test2     

for classifier, vectori in test:
    _vectorizer = vectori
    _xtrain = _vectorizer.fit_transform(tokens)

    xtrain, xvalid, ytrain, yvalid = train_test_split(_xtrain, df['cent'], random_state=42, test_size=0.2)

    classi = classifier
    classi = classi.fit(xtrain, ytrain) # xtrain_bow:bag of words features for train data, ytrain: train data labels
    prediction = classi.predict(xvalid) #predict on the validation set
    print("classifier= "+classifier.__class__.__qualname__ +"\tvectorizer= "+vectori.__class__.__qualname__+"\tscore= ",end=" ")
    print(f1_score(yvalid, prediction,average='macro'))#evaluate on the validation set
    
#filtered_tweets 1000 features
# classifier= KNeighborsClassifier   vectorizer= CountVectorizer score=  0.3948041779324909
# classifier= LogisticRegression     vectorizer= CountVectorizer score=  0.5480840509487324
# classifier= RandomForestClassifier vectorizer= CountVectorizer score=  0.5412853916561066
# classifier= SVC                    vectorizer= CountVectorizer score=  0.5560747304719477
# classifier= XGBClassifier          vectorizer= CountVectorizer score=  0.4682263619864087
# classifier= KNeighborsClassifier   vectorizer= TfidfVectorizer score=  0.3855962473051861
# classifier= LogisticRegression     vectorizer= TfidfVectorizer score=  0.5410924449938832
# classifier= RandomForestClassifier vectorizer= TfidfVectorizer score=  0.5402665150162561
# classifier= SVC                    vectorizer= TfidfVectorizer score=  0.5568693739232171
# classifier= XGBClassifier          vectorizer= TfidfVectorizer score=  0.4589895909384849

#lemmatized_tweets 1000 features
# classifier= KNeighborsClassifier   vectorizer= CountVectorizer score=  0.3948041779324909
# classifier= LogisticRegression     vectorizer= CountVectorizer score=  0.5480840509487324
# classifier= RandomForestClassifier vectorizer= CountVectorizer score=  0.5390131130387873
# classifier= SVC                    vectorizer= CountVectorizer score=  0.5560747304719477
# classifier= XGBClassifier          vectorizer= CountVectorizer score=  0.4682263619864087
# classifier= KNeighborsClassifier   vectorizer= TfidfVectorizer score=  0.3855962473051861
# classifier= LogisticRegression     vectorizer= TfidfVectorizer score=  0.5410924449938832
# classifier= RandomForestClassifier vectorizer= TfidfVectorizer score=  0.5399410627604814
# classifier= SVC                    vectorizer= TfidfVectorizer score=  0.5568693739232171
# classifier= XGBClassifier          vectorizer= TfidfVectorizer score=  0.4589895909384849

#lemmatized_tweets 3000 features
# classifier= KNeighborsClassifier	vectorizer= CountVectorizer	score=  0.3973835724294665
# classifier= LogisticRegression	vectorizer= CountVectorizer	score=  0.5633900668321052
# classifier= RandomForestClassifier	vectorizer= CountVectorizer	score=  0.5663547012446637
# classifier= SVC	vectorizer= CountVectorizer	score=  0.5592635330007211
# classifier= XGBClassifier	vectorizer= CountVectorizer	score=  0.4561932870568634
# classifier= KNeighborsClassifier	vectorizer= TfidfVectorizer	score=  0.3433009865555376
# classifier= LogisticRegression	vectorizer= TfidfVectorizer	score=  0.5605646507506697
# classifier= RandomForestClassifier	vectorizer= TfidfVectorizer	score=  0.558489402709832
# classifier= SVC	vectorizer= TfidfVectorizer	score=  0.5793119775200134
# classifier= XGBClassifier	vectorizer= TfidfVectorizer	score=  0.45522587677834997
```
