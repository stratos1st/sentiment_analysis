```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer

# read tokens
infile = open('filtered_tweets.pickle','rb')
tokens = pickle.load(infile)
infile.close()

# # concatinate tokens. ending up with a filtered tweet
tok=[]
for tweet in tokens:
    tmp=''
    for word in tweet:
        tmp=tmp+' '+word
    tok.append(tmp)
tokens=tok

tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf = tfidf_vectorizer.fit_transform(tokens)

print(tfidf)
print(tfidf.toarray())
print(tfidf_vectorizer.get_feature_names()) 
print(tfidf.shape) #the output is a numpy array of features
```
